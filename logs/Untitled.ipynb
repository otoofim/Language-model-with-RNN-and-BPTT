```python
import numpy
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
errors=[]
with open("result.txt", "r") as myfile:
    for line in myfile:
        line=line.strip()
        to=line.split(" ")
        errors.append(-1*float(to[4].replace(",","")))
print(max(errors))
```

```python
x=range(0,len(errors))
fig = plt.figure()
plt.plot(x,errors,'-go')
plt.ylabel("Error")
plt.xlabel("")
#plt.legend(loc='best')
plt.show()
fig.savefig('temp.png', dpi=1000)
```
